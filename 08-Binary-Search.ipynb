* Import Statement
* Creating a list --> unique and orderd slightly random list --> near to real world example

```
from random import randint
random_set = set()
for i in range(-1000, 1001, 5):
    sort_set = randint(0,1000)
    random_set.add(sort_set)

random_list = sorted(list(random_set))
random_list
```

Defining the function for naive or normal search, mostly this is for loop thing

```
def normal_search(find, sorted_list):
    for i, v in enumerate(sorted_list):
        if v == find:
            return i

print(normal_search(995, random_list))
len(random_list)
```

* Defining the binary search function

```
l = 0
h = len(random_list) - 1

def binary_search(high, low, sorted_list, find):

    mid_value = (low + high) // 2
    if high >= low:

        if find == sorted_list[mid_value]:
            return mid_value

        elif find > sorted_list[mid_value]:
            return binary_search(high,mid_value + 1, sorted_list, find)

        elif find < sorted_list[mid_value]:
            return binary_search(mid_value -1 , low, sorted_list, find)
    else:
        return f"OOP's Your element is not present in the memory"

binary_search(h, l, random_list, 995)
```

* Using time module to check the time

```
from time import time

beginning1 = time()
print(normal_search(995, random_list))
ending1 = time()
print("Time taken for Normal Search is ", ending1 - beginning1)

beginning = time()
print(binary_search(h, l, random_list, 995))
ending = time()
print("Time taken for binary search is ", ending - beginning)
```

```
from random import randint
from time import time

random_set = set()
for i in range(-1000, 10001, 5):
    sort_set = randint(0,10000)
    random_set.add(sort_set)
random_list = sorted(list(random_set))


l = 0
h = len(random_list) - 1
def binary_search(high, low, sorted_list, find):

    mid_value = (low + high) // 2
    if high >= low:

        if find == sorted_list[mid_value]:
            return mid_value

        elif find > sorted_list[mid_value]:
            return binary_search(high,mid_value + 1, sorted_list, find)

        elif find < sorted_list[mid_value]:
            return binary_search(mid_value -1 , low, sorted_list, find)
    else:
        return f"OOP's Your element is not present in the memory"


def normal_search(find, sorted_list):
    for i, v in enumerate(sorted_list):
        if v == find:
            return i
        else:
            return "OOP's your element is not present in the memory"

find_number = int(input("Enter your number to search between 0 to 10,000: "))
print("\n")

beginning1 = time()
print(normal_search(find_number, random_list))
ending1 = time()
normal_time = ending1 - beginning1
print(f"Time taken for 'Normal Search' is {normal_time} \n")

beginning = time()
print(binary_search(h, l, random_list, find_number))
ending = time()
binary_time = ending - beginning
print(f"Time taken for 'Binary search' is {binary_time} \n")

diff = normal_time - binary_time
print(f"{diff} seconds differnce between binary search and normal search \n")

if diff < 0:
    print("You number is present beginning of random number list, so it normal or naive search finds that easily, try entering large number ")
else:
    print("See I told you, binary search is faster when your number is present last in the list")
```
